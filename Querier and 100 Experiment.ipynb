# Do experiement 100 times

```python
import time

# Compressed Index
cmp_start = time.time()
for i in range(100):
    res = Querier.do_query(i, compressed = True)
cmp_end = time.time()


ucmp_start = time.time()
for i in range(100):
    res = Querier.do_query(i, compressed = False)
ucmp_end = time.time()
```

```python
print("Compressed Indexing used time: {}".format(cmp_end - cmp_start))
print("Uncompressed Indexing used time: {}".format(ucmp_end - ucmp_start))
```

# Conclusion
- The compressed data querying is faster. Let alone that reading compressed data need to fetch more metadata like offset and size.
